## ntlk 
- 가장 오래된 자연어 처리 라이브러리
- 한국어 지원 X

```python
import nltk 
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
sentence= """
At eight o'clock on Thursday morning
Arthur didn't feel very good."""
sentence
```

- 토큰화

```python
tokens = nltk.word_tokenize(sentence)
tokens
```

- 문장속 단어에 품사 적용

```python
tagged = nltk.pos_tag(tokens)
tagged
```

```python
# 동사, 명사만 추출해서 리스트에 append 
[token for token, pos in tagged if pos.startswith("N") or pos.startswith("V")]
```

## spacy
- python 기반 라이브러리
- 다양한 언어 지원 

```python
import spacy
from spacy.lang.en.examples import sentences
```

- 품사 구분 외에도 표현해주는 기능이 존재

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp(sentences[0])
print(doc.text)
print('-' * 80)
print("단어","원형","품사","태그", "의존성", "모양", "알파벳", "금칙어",sep="\t")
for token in doc:
    print(
        token.text # 단어
        , token.lemma_ # 원형
        , token.pos_ # 품사
        , token.tag_ # 태그
        , token.dep_ # 의존성
        , token.shape_ # 모양
        , token.is_alpha # 알파벳
        , token.is_stop # 금칙어
        , sep='\t')
```

### Okt 

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
txt = '아버지가방에들어가신다.'
```

```python
okt.pos(txt)
```

```python
[token[0] for token in okt.pos(txt) if token[1][0] in 'NVJ']
```

```python
okt.nouns(txt)
```
