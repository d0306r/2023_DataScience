## 데이터 수집

```python
from sklearn.datasets import load_breast_cancer

# 사이킷런 데이터 객체 
breast_cancer = load_breast_cancer()
```

```python
# 타겟 데이터가 범주형이므로 분류 모델을 사용해야 한다는 것을 알 수 있다. 
breast_cancer.target_names
```

```python
# description의 약자, 기본 정보를 보여준다. 
print(breast_cancer.DESCR)
```

## 데이터 점검 및 탐색

```python
import numpy as np
import pandas as pd
```

```python
# 데이터 프레임 컬럼 값을 featrue로 설정 
df_cancer = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)

# target을 끌어와 데이터 프레임에 컬럼 생성
df_cancer['target'] = breast_cancer['target']
df_cancer.info()
```

```python
# target data가 확실히 분류의 형태를 띄고 있다는 것을 알 수 있다. 
df_cancer.describe()
```

## 전처리 및 정제

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# standardization
scaler = StandardScaler()

# 라벨 값을 제거하고 feature만 스케일링을 한다. 
scaled = scaler.fit_transform(df_cancer.drop('target', axis=1))
```

```python
# target이 빠져 column이 한 개 제거된 것을 확인할 수 있다. 
print(f'df_cancer.shape: {df_cancer.shape}')
print('-'*50)
print(f'scaled.shape : {scaled.shape}')
```

```python
scaled[:1, :]
```

```python
scaled[5:7, :]
```

```python
# train은 학습할때, valid는 평가할 때 사용 
x_train, x_valid, y_train, y_valid = train_test_split(scaled, df_cancer['target'], random_state = 10)
```

```python
# y는 정답이므로 row가 1개만 존재한다. 
x_train.shape, y_train.shape
```

```python
# train data set이 훨씬 많음을 알 수 있다. 그러나 row의 개수는 일치해야 한다. 
x_valid.shape, y_valid.shape
```

## 모델링 및 훈련

```python
from sklearn.linear_model import LogisticRegression
```

```python
# 로지스틱 회귀 모델 선택 
lr_clf = LogisticRegression()
```

```python
# train data로 학습 및 훈련 
lr_clf.fit(x_train, y_train)
```

## 평가

```python
from sklearn.metrics import accuracy_score
```

```python
# 훈련이 끝난 lr_clf 변수로 예측 
pred = lr_clf.predict(x_valid)
```

```python
# 정확도 측정
accuracy_score(y_valid, pred)
```
