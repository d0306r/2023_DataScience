```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

### matplotlib 기초

```python
plt.figure() # 그래프를 그리기 위한 툴 
plt.axes() # 좌표 평면
plt.show() # 화면에 표시
```

### 단일 그래프
plot
- line plot을 그려주는 함수
- figure, axes가 없으면, 이를 만들어 그려준다.

```python
plt.plot(np.arange(2,7))
plt.show()
```

### 다중 그래프
- 1개의 axis 안에 여러가지 그래프를 그린다.

```python
# 틀 크기 (너비, 높이)
plt.figure(figsize=[15,5])

# 그래프 2개 그리기
plt.plot(np.arange(5))
plt.plot(np.arange(2,7))

# 화면에 표시
plt.show()
```

### Subplots


```python
fig, ax = plt.subplots(2,2, figsize=[15,5])
print(ax.shape) # (열, 행)의 형태로 표현 

ax[0,0].plot(np.arange(5)) # 첫번째 왼쪽
ax[0,1].plot(np.arange(5)) # 첫번째 오른쪽
ax[0,1].plot(np.arange(2,6)) # 첫번째 오른쪽
ax[1,0].plot(np.arange(5)) # 두번째 왼쪽
ax[1,1].plot(np.arange(5)) # 두번째 오른쪽 

plt.show()
```

```python
arr = (np.arange(5))
arr1 = (np.arange(2,6))
plt.figure(figsize=[15,5])

plt.subplot(2,2,1) # (2,2) shape의 1번째 자리에 넣을 그래프라는 의미 
plt.plot(arr)

plt.subplot(2,2,2)
plt.plot(arr)
plt.plot(arr1)

plt.subplot(2,2,3)
plt.plot(arr)

plt.subplot(2,2,4)
plt.plot(arr)

plt.show()
```

```python
arr = (np.arange(5))
arr1 = (np.arange(2,6))
arr2 = np.arange(1,5)
plt.figure(figsize=[15,5])

plt.subplot(2,3,1) # (2,2) shape의 1번째 자리에 넣을 그래프라는 의미 
plt.plot(arr)

plt.subplot(2,3,2)
plt.plot(arr)
plt.plot(arr1)

plt.subplot(2,3,3)
plt.plot(arr)

plt.subplot(2,3,4)
plt.plot(arr)

plt.subplot(2,3,5)
plt.plot(arr)
plt.plot(arr2)

plt.subplot(2,3,6)
plt.plot(arr2)

plt.show()
```

```python
# for문 사용해서 그래프 그리기
for i in range(1, (ax.shape[0]*ax.shape[1])+1) :
    arr = np.arange(i,i+5)
    plt.subplot(ax.shape[0], ax.shape[1], i)
    plt.plot(arr)
plt.show()
```

### 타이틀
- figure의 제목 : suptitle("제목")

```python
fig, ax = plt.subplots(1,2,figsize=[15,5])
fig.suptitle("figure title")

x = range(0,10)
y = np.exp(x) # 지수함수
ax[0].plot(x,y)

x = range(1,1000)
y = np.log(x) # 로그함수 
ax[1].plot(x,y)

plt.show()
```
