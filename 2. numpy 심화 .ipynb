### 선형대수(벡터/행렬연산) 함수

- L1 Norm : 스칼라들에 절대값을 취한 뒤 합한 것 
- L2 Norm : 스칼라들을 각각 제곱해서 더한 후 루트를 씌운 것, 제곱을 하기 때문에 오차에 더 민감해서, L1보다 많이 사용됨 

  Outlier 가 적당히 무시되길 원한다면 L1 Loss 를 사용, Outlier 의 등장에 신경써야 하는 경우라면 L2 Loss 를 사용

image.png

- L1 distance (Manhattan) : 두 개의 고정된 벡터 사이에 여러개의 path 발생
- L2 distance (Euclidean) : 하나의 unique shortest path만이 존재 

```python
u = np.array([1, -2, 3, -4, 5])
```

```python
def l1_norm(x) :
  x_norm = np.abs(x) # 절대값
  return np.sum(x_norm)

l1_norm(u)
```

```python
def l2_norm(x) :
  x_norm = np.sum(x*x) # 스칼라의 제곱 
  return np.sqrt(x_norm) # 제곱값의 루트 

l2_norm(u)
```

```python
np.linalg.norm(u,1) # 벡터 정규화 함수, L1은 상대적으로 적게 사용 
```

```python
np.linalg.norm(u) # L2 norm 
```

### 벡터연산(내적)

행렬 x 행렬
- 조건
  - 두 벡터의 차원(길이)이 같아야 함 
  - 앞 벡터는 행, 뒷 벡터는 열이어야 함 
- 특징
  - 두 벡터 각 요소끼리의 곱의 합
  - 결과 값은 스칼라

```python
# 1차원 배열 내적
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
```

```python
x @ y # 벡터 내적 연산산
```

```python
np.dot(x,y) 
```

```python
x.dot(y.T) # y의 축을 바꿈 
```

```python
y.dot(x.T) # x의 축을 바꿈
```

2차원 배열 내역

- 계산 이후 shape는 같은 숫자를 날린 값과 같음

```python
np.dot(x.T, y)
```

### 역행렬
행렬 A와 B를 곱했을 때 단위행렬 I와 같다면, B를 A의 역행렬이라고 함 

```python
from numpy.linalg import inv 
```

```python
x = np.array(([[1,2,3],[1,0,0],[0,0,1]]))
```

```python
y = inv(x)
y
```

```python
x.dot(y) # 단위행렬 생성, 따라서 y는 x의 역행렬 
```

#### newaxis
차원 추가

```python
x = np.arange(3) # 1차원
print(x.shape)
x
```

```python
x_na = x[:, np.newaxis] # 2차원, 열에 리스트 할당 
print(x_na.shape)
x_na
```

```python
x_na1 = x[np.newaxis, :] # 행에 리스트 할당 
x_na1
```

```python
x_na_na = x[:, np.newaxis, np.newaxis] # 3차원, 층에 2차원 배열 할당  
print(x_na_na.shape)
x_na_na
```
